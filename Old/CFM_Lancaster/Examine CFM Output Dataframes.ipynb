```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
results_path = f'/home/jez/Community_Firn_Model/results/CFM_output/'
# results_path = f'/home/jez/DSNE_ice_sheets/Community_Firn_Model/Processed_Data/CFM_Output_Data/'
ds_single_dimension_outputs = xr.open_dataset(f"{results_path}Single_Dimension_CFM_Output.nc")

stations = ['Summit','Kan-M']
ds_multi_dimension_list = []
for station in stations:
    results_folder = f"{results_path}{station}/"
    ds = xr.open_dataset(f"{results_folder}Reformatted_Results_{station}.nc")
    ds_multi_dimension_list.append(ds)
```

```python
stations = ['Summit','Kan-M']
outputs = ['delta_ele','cum_ele','tot_compaction','cum_tot_compaction','depth','density']
outputs_fullnames = ['Elevation Change / m','Cumulative Elevation Change / m','Total Compaction / m','Cumulative Total Compaction / m']

fig, subplots = plt.subplots(2, 2,figsize=(14,8))
for station in stations:
    for ax,output in zip(subplots.flat,outputs):
        ds_single_dimension_outputs[output].isel(time=slice(1,None)).sel(station=station).plot.line(x='year',alpha=0.6,ax=ax,label=station)
        ax.set_title('')
        ax.legend()
fig.tight_layout()
```

```python
ds = ds_multi_dimension_list[1].isel(time=slice(1,None))
plt.figure(figsize=(12,12))
ds.isel(time=slice(None,500,100) , cell=slice(None,None,10)).plot.scatter(y='depth',x='density',hue='year',hue_style='discreet',marker='x')
plt.title('Kan-M Depth-Density Profile at Different Times')
plt.show()
```
